```python
import numpy as np
import matplotlib.pyplot as plt

#Estilo para la gráfica
%matplotlib inline
plt.style.use('seaborn-v0_8-darkgrid')

```

```python
# ============== CASO PRÁCTICO 01 ==============

def caso_practico_01():
    lista1 = [1, 2, 3, 4, -4, -5, -6, 23, 25, 27, 100, 200, 240, 240]
    lista2 = [-4, -5, -6, 300, 400, 600, 500, 500]

    conjunto1 = set(lista1)
    conjunto2 = set(lista2)

    print("\n" + "=" * 50)
    print("CASO PRÁCTICO 01 - TEORÍA DE CONJUNTOS")
    print("=" * 50)
    print(f"\nLista 1 original: {lista1}")
    print(f"Lista 2 original: {lista2}")
    print("\nTrabajaremos sin duplicados (vista como conjunto):")
    print(f"Conjunto 1: {sorted(conjunto1)}")
    print(f"Conjunto 2: {sorted(conjunto2)}")

    while True:
        print("\n--- Submenú Caso Práctico 01 ---")
        print("1. Cardinalidad 1")
        print("2. Cardinalidad 2")
        print("3. Unión")
        print("4. Diferencia 1 (A - B)")
        print("5. Diferencia 2 (B - A)")
        print("6. Intersección")
        print("0. Volver al menú principal")

        opcion = input("\nSeleccione una opción: ")

        if opcion == "1":
            print(f"\n-> Cardinalidad del Conjunto 1: {len(conjunto1)} elementos")
        elif opcion == "2":
            print(f"\n-> Cardinalidad del Conjunto 2: {len(conjunto2)} elementos")
        elif opcion == "3":
            union = conjunto1.union(conjunto2)
            print(f"\n-> Unión (A ∪ B): {sorted(union)}")
            print(f"  Total: {len(union)} elementos")
        elif opcion == "4":
            diferencia1 = conjunto1.difference(conjunto2)
            print(f"\n-> Diferencia 1 (A - B): {sorted(diferencia1)}")
            print(f"  Total: {len(diferencia1)} elementos")
        elif opcion == "5":
            diferencia2 = conjunto2.difference(conjunto1)
            print(f"\n-> Diferencia 2 (B - A): {sorted(diferencia2)}")
            print(f"  Total: {len(diferencia2)} elementos")
        elif opcion == "6":
            interseccion = conjunto1.intersection(conjunto2)
            print(f"\n-> Intersección (A ∩ B): {sorted(interseccion)}")
            print(f"  Total: {len(interseccion)} elementos")
        elif opcion == "0":
            break
        else:
            print("\nOpción inválida. Intente nuevamente.")


# ============== CASO PRÁCTICO 02 ==============


def solicitar_flotante(mensaje: str):
    try:
        return float(input(mensaje))
    except ValueError:
        print("Valor inválido.")
        return None


def funcion_una_variable(x: float) -> float:
    """f(x) = x^2 + 3"""
    return x**2 + 3


def funcion_dos_variables(x: float, y: float) -> float:
    """f(x, y) = x^2 + y^2"""
    return x**2 + y**2


def f(x: float) -> float:
    """f(x) = 2x + 1"""
    return 2 * x + 1


def g(x: float) -> float:
    """g(x) = x^2"""
    return x**2


def funcion_compuesta(x: float) -> float:
    """Función compuesta f(g(x))"""
    return f(g(x))


def funcion_graficar(x):
    """f(x) = x^2 + 2"""
    return x**2 + 2


def caso_practico_02():
    print("\n" + "=" * 50)
    print("CASO PRÁCTICO 02 - FUNCIONES")
    print("=" * 50)

    while True:
        print("\n--- Submenú Caso Práctico 02 ---")
        print("1. Función de una sola variable: f(x) = x^2 + 3")
        print("2. Función de dos variables: f(x,y) = x^2 + y^2")
        print("3. Función compuesta: f(g(x)) donde f(x)=2x+1, g(x)=x^2")
        print("4. Gráfica de función: f(x) = x^2 + 2")
        print("0. Volver al menú principal")

        opcion = input("\nSeleccione una opción: ")

        if opcion == "1":
            print("\nFunción de una variable: f(x) = x^2 + 3")
            print("\nEjemplos predefinidos:")
            print(f"f(-5) = {funcion_una_variable(-5)}")
            print(f"f(-1) = {funcion_una_variable(-1)}")
            print(f"f(4) + f(3) = {funcion_una_variable(4)} + {funcion_una_variable(3)} = {funcion_una_variable(4) + funcion_una_variable(3)}")

            respuesta = input("\n¿Desea evaluar con valores propios? (s/n): ")
            if respuesta.lower() == 's':
                x = solicitar_flotante("Ingrese el valor de x: ")
                if x is not None:
                    print(f"f({x}) = {funcion_una_variable(x)}")

        elif opcion == "2":
            print("\nFunción de dos variables: f(x,y) = x^2 + y^2")
            print("\nEjemplos predefinidos:")
            print(f"f(2, 3) = {funcion_dos_variables(2, 3)}")
            print(f"f(5, -3) = {funcion_dos_variables(5, -3)}")

            respuesta = input("\n¿Desea evaluar con valores propios? (s/n): ")
            if respuesta.lower() == 's':
                x = solicitar_flotante("Ingrese el valor de x: ")
                y = solicitar_flotante("Ingrese el valor de y: ")
                if x is not None and y is not None:
                    print(f"f({x}, {y}) = {funcion_dos_variables(x, y)}")

        elif opcion == "3":
            print("\nFunción compuesta: f(g(x))")
            print(" donde f(x) = 2x + 1 y g(x) = x^2")
            print("\nEjemplos predefinidos:")
            print(f"f(g(-5)) = f({g(-5)}) = {funcion_compuesta(-5)}")
            print(f"f(g(-2)) = f({g(-2)}) = {funcion_compuesta(-2)}")

            respuesta = input("\n¿Desea evaluar con valores propios? (s/n): ")
            if respuesta.lower() == 's':
                x = solicitar_flotante("Ingrese el valor de x: ")
                if x is not None:
                    print(f"g({x}) = {g(x)}")
                    print(f"f(g({x})) = {funcion_compuesta(x)}")

        elif opcion == "4":
            print("\nGraficando función: f(x) = x^2 + 2")
            x = np.linspace(-10, 10, 400)
            y = funcion_graficar(x)

            plt.figure(figsize=(5, 3))
            plt.plot(x, y, color='#2E86AB', linewidth=3, alpha=0.9)
            plt.axhline(y=0, color='#A23B72', linewidth=1.5, alpha=0.7)
            plt.plot(0, 2, 'ro', markersize=10)
            plt.xlabel('x', fontsize=12)
            plt.ylabel('f(x)', fontsize=12)
            plt.title('Gráfica de f(x) = x^2 + 2', fontsize=14, fontweight='bold')
            plt.tight_layout()
            plt.show()
            print("Gráfica mostrada correctamente.")

        elif opcion == "0":
            break
        else:
            print("\nOpción inválida. Intente nuevamente.")


# ============== MENÚ PRINCIPAL ==============


# Punto de entrada del menú principal para navegar por los casos
def menu_principal():
    print("\n" + "=" * 60)
    print(" " * 8 + "TEORÍA DE CONJUNTOS, RELACIONES Y FUNCIONES")
    print("=" * 60)

    while True:
        print("\n--- MENÚ PRINCIPAL ---")
        print("1. Caso Práctico 01 - Operaciones con Conjuntos")
        print("2. Caso Práctico 02 - Evaluación de Funciones")
        print("0. Salir")

        opcion = input("\nSeleccione una opción: ")

        if opcion == "1":
            caso_practico_01()
        elif opcion == "2":
            caso_practico_02()
        elif opcion == "0":
            print("\n¡Gracias por usar el programa! Hasta pronto.\n")
            break
        else:
            print("\nOpción inválida. Intente nuevamente.")
```

```python
# Punto de entrada del menú principal
menu_principal()
```
